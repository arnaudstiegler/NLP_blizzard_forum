```python
import json

with open('data/10000_posts.json','r') as file:
    data = json.load(file)
#Creating the csv file for gtp2 fine-tuning

raw_comments = []

for elem in data:
    raw_comments += elem[1]
    
comments_df = pd.DataFrame(raw_comments)
comments_df.to_csv('data/gpt2_1000_comments.csv')
```
